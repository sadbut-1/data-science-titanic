# Introdução

O relatório via analisar os dados dos passageiros que estavam no acidente do Titanic. A princípio foi necessário um tratamento nos dados para completar, excluir ou ajustar certos valores incompletos ou incorretos. Depois foram levantadas questões relevantes para tentar descobrir um pouco mais sobre os perfis dos passageiros e porque alguns grupos tiveram um número maior de sobreviventes que outros.

As questões respondidas foram:
1. Quantos sobreviveram no acidente?
2. Quantos morreram no acidente?
3. Qual a relação entre os os sobreviventes e os pontos de embarque?
4. Qual a relação entre as tarifas e os sobreviventes?
5. Qual a relação dos sobreviventes com a classe de embarque?
6. Qual a proporção de sobreviventes por sexo?
7. Qual faixa de idade teve mais sobreviventes?
8. Qual a relação de pessoas com familiares e os sobreviventes?

O conjunto de dados utilizado foi o titanic-data-6.csv https://d17h27t6h515a5.cloudfront.net/topher/2017/October/59e4fe3d_titanic-data-6/titanic-data-6.csv

# Preparação dos dados

```python
#Import das bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#Lendo arquivo do Titanic
df = pd.read_csv('titanic-data-6.csv')
```

```python
# Mostrando a estrutura da tabela
df.head()
```

```python
# Verificando informações dos dados
df.info()
```

- Verificando os valores faltantes do conjunto de dados.

```python
df.isnull().sum()
```

```python

def cleaned_data(df):
    """Prepara os dados para serem analizados
       Argumentos: 
           df: DataFrame
       Retorno
           retorn um DataFrame
    """
    # Colocando os indices em minusculo pra facilitar o trabalho de análise
    df.rename(columns=lambda x: x.lower(), inplace=True)
    
    # Completando coluna Embarked
    df['embarked'].fillna(df['embarked'].mode()[0], inplace = True)
    
    # Completando as idades faltantes com a média de idade
    df['age'].fillna((df['age'].mean()), inplace = True)
    
    # Esta faltando muitos valores na coluna Cabin, por isso será descartada
    df.drop(['cabin'], axis=1, inplace = True)
    
    return df
```

```python
#Conferindo novamente
df_clean = cleaned_data(df)

df_clean.isnull().sum()
```

Na preparação dos dados, foi verificado a ausencia de alguns valores em determinadas colunas. Diante disso foram feitos os seguintes ajustes:
- Na coluna Embarked faltavam apenas 2 valores e foram completados utilizando mode.
- Na coluna idade faltavam mais valores (177). Por considerar este dado importante, ao invés de descatar a coluna, os dados nulos foram preechidos com a média de idade.
- Na coluna Cabin estava faltando mais da metade dos valores e por isso a coluna foi descartada.

# Questões

**1. Quantos sobreviveram no acidente?**

```python
def count_survived(sved):
    """Retorna o numero e o porcentual de sobreviventes
       Argumentos: 
           sved: Integer
       Retorno
           Retorna uma lista com 2 inteiros
    """
    total_passageiros = 891
    sobreviventes = df_clean.query('survived == @sved')['survived'].count()
    percentual = ((sobreviventes * 100)/total_passageiros).astype(int)
    return [sobreviventes, percentual]
```

```python
print('Dos 891 passageiros listados, {} sobreviveram ({}%).'.format(count_survived(1)[0], count_survived(1)[1]))
```

**2. Quantos morreram no acidente?**

```python
print('Dos 891 passageiros listados, {} não sobreviveram({}%).'.format(count_survived(0)[0], count_survived(0)[1]))
```

**3.  Qual a relação entre os os sobreviventes e os pontos de embarque ?**

- Como verificado acima, faltam alguns dados sobre o embarque. Resolvi preenchelos utilizando a função mode()

```python
def embarked_by_port(data, port):
    """Retorna a média de sobreviventes por ponto de embarque
       Argumentos: 
           data: DataFrame
           port: String
       Retorno
           Retorna um Float
    """
    return data.query('embarked == @port')['survived'].mean()

s = embarked_by_port(df_clean, 'S')
c = embarked_by_port(df_clean, 'C')
q = embarked_by_port(df_clean, 'Q')

plt.axis('equal') 
plt.title('Relação Sobreviventes x Pontos de embarque')
plt.pie([s, c, q], labels=['Southampton', 'Cherbourg', 'Queenstown'], shadow=True);
```

- Cherbourg teve o maior número de sobreviventes. Mas seria algo aleatório ou por algum outro fator?


**4. Qual a relação entre as tarifas e os sobreviventes ?**

```python
#Verificando a média do valor pago agrupado por sobreviventes.
df_fare = df_clean.groupby('survived')['fare'].mean()

plt.bar([1, 2], [df_fare[0], df_fare[1]], tick_label=['Pagou menos', 'Pagou mais']);
plt.title("Relação entre valor das tarifas e sobreviventes")
plt.ylabel('Sobreviventes')
```

- Vemos acima que passageiros que pagaram mais caro em suas passagens tiveram um numero maior de sobreviventes

```python
def survived_fare(data, port):
    """Retorna o valor medio do bilhete por porto
       Argumentos: 
           data: DataFrame
           port: String
       Retorno
           Retorna um Float
    """
    return data.query('embarked == @port')['fare'].mean()

s = survived_fare(df_clean, 'S')
c = survived_fare(df_clean, 'C')
q = survived_fare(df_clean, 'Q')

barlist = plt.bar([1, 2, 3], [s, c, q], tick_label=['Southampton', 'Cherbourg', 'Queenstown']);
barlist[0].set_color('y')
barlist[1].set_color('g')
barlist[2].set_color('r')
plt.ylabel('Média do valor da passagem')
```

- Aqui verificamos que os passageiros que paragaram mais caro embarcaram principalmente em Cherbourg

```python
g = sns.factorplot("pclass", col="embarked",  data=df,
                   size=6, kind="count", palette="muted")
g.despine(left=True)
g = g.set_ylabels("Quantidade")
```

- Não da pra saber o motivo da sobrevivencia, mas posso concluir que passageiros da primeira classe tiveram um numero maior de sobreviventes. Talvez pela localização de suas cabines.

**5. Qual a relação dos sobreviventes com a classe de embarque?**

```python
def survived_class(pclass):
    """Retorna o numero e o porcentual de sobreviventes por porto
       Argumentos: 
           pclass: String
       Retorno
           Retorna uma lista com 2 inteiros
    """
    total_classe = df_clean.query('pclass == @pclass')['pclass'].count()
    sobreviventes = df_clean.query('survived == 1').query('pclass == @pclass')['survived'].count()
    porcentual = ((sobreviventes * 100)/total_classe).astype(int)
    return [sobreviventes, porcentual]

c1 = survived_class(1)
c2 = survived_class(2)
c3 = survived_class(3)

plt.rcdefaults()
fig, ax = plt.subplots()

classe = ('1', '2', '3')
y_pos = np.arange(len(classe))
sobreviventes = [c1[1], c2[1], c3[1]]

ax.barh(y_pos, sobreviventes, align='center',
        color='green', ecolor='black')
ax.set_yticks(y_pos)
ax.set_yticklabels(classe)
ax.invert_yaxis()  
ax.set_xlabel('Sobreviventes')
ax.set_ylabel('Classes')
ax.set_title('Relação dos sobreviventes com a classe em %')

plt.show()

print('Primeira classe: {}%'.format(c1[1]))
print('Segunda classe: {}%'.format(c2[1]))
print('Terceira classe: {}%'.format(c3[1]))
```

**6. Qual a proporção de sobreviventes por sexo?**

```python
def get_survived_by_sex(df, sex, sved = 1):
    """Retorna o numero de sobreviventes por sexo
       Argumentos: 
           df: DataFrame
           sex: String
           sved: Integer
       Retorno
           Retorna um DataFrame
    """
    return df.query('sex == @sex').query('survived == @sved')
```

```python
female_d = get_survived_by_sex(df_clean, 'female', 0)
female_s = get_survived_by_sex(df_clean, 'female')
male_d = get_survived_by_sex(df_clean, 'male', 0)
male_s = get_survived_by_sex(df_clean, 'male')

porcentual_female = ((female_s.shape[0] * 100)/count_survived(1)[0]).astype(int)
porcentual_male = ((male_s.shape[0] * 100)/count_survived(1)[0]).astype(int)
print('Sobreviventes do sexo feminino: {}%'.format(porcentual_female))
print('Sobreviventes do sexo masculino: {}%'.format(porcentual_male))
```

```python
N = 2
morreram = (male_d.shape[0], female_d.shape[0])
sobreviveram = (male_s.shape[0], female_s.shape[0])
ind = np.arange(N)
width = 0.45 

p1 = plt.bar(ind, morreram, width)
p2 = plt.bar(ind, sobreviveram, width, bottom=morreram)

plt.title('Proporção de sobreviventes por sexo')
plt.xticks(ind, ('Homens', 'Mulheres'))
plt.ylabel('Numero de passageiros')
plt.legend((p1[0], p2[0]), ('Mortos', 'Sobreviventes'))

plt.show()
```

- Proporcionalmente, mulheres tiveram um número muito maior de sobreviventes.

**7. Qual faixa de idade teve mais sobreviventes?**

```python
g = sns.kdeplot(df["age"][(df["survived"] == 0)], color="Red", shade = True)
g = sns.kdeplot(df["age"][(df["survived"] == 1)], ax =g, color="Blue", shade= True)
g.set_xlabel("Idade")
g.set_ylabel("Ocorrências")
g.set_title('Faixa de idade dos sobreviventes')
g = g.legend(["Não Sobreviventes","Sobreviventes"])
```

- Pelo gráico acima, pode-se observar que crianças (0 a +- 10 anos) tiveram um número grande de sobreviventes.

##### Verificando os sobreviventes por idade e sexo
- É interessante observar nos 2 gráficos abaixo, que a maior parte das crianças sobreviventes era do sexo masculino. Por outro lado, sobreviventes do sexo feminino eram em sua maior parte adultos.

```python
g = sns.kdeplot(female_d["age"], color="Red", shade = True)
g = sns.kdeplot(female_s["age"], ax =g, color="Blue", shade= True)
g.set_xlabel("Idade")
g.set_ylabel("Ocorrências")
g.set_title("Faixa de idade dos sobreviventes do sexo feminino")
g = g.legend(["Não Sobreviventes","Sobreviventes"])
```

- No gráfico acima foram cruzados a idade e o sexo para verificar a faixa de idade onde houve mais sobreviventes  do sexo feminimo 

```python
g = sns.kdeplot(male_d["age"], color="Red", shade = True)
g = sns.kdeplot(male_s["age"], ax =g, color="Blue", shade= True)
g.set_xlabel("Idade")
g.set_ylabel("Ocorrências")
g.set_title("Faixa de idade dos sobreviventes do sexo masculino")
g = g.legend(["Não Sobreviventes","Sobreviventes"])
```

- No gráfico acima foram cruzados a idade e o sexo para verificar a faixa de idade onde houve mais sobreviventes  do sexo masculino

**8. Qual a relação de pessoas com familiares e os sobreviventes?**

```python
group_1 = df_clean.query('sibsp > 0 | parch > 0')
group_2 = df_clean.query('sibsp == 0 & parch == 0')

group_1['alone'] = np.repeat(0, group_1.shape[0])
group_2['alone'] = np.repeat(1, group_2.shape[0])

df_group = group_1.append(group_2)
df_alone = df_group.groupby('alone')['survived'].mean()

barlist = plt.bar([1, 2], [df_alone[0], df_alone[1]], tick_label=['Familia', 'Sozinhos']);
barlist[0].set_color('y')
plt.ylabel('Sobreviventes')
```

- No gráfico acima verificamos que pessoas acompanhadas de algum familiar tiveram mais chance de sobrevivencia.

# Conclusões

Analisando os dados do acidente do Titanic, fica claro que os passageiros das classes superiores foram previlegiados no momento de embarcarem nos botes salva vidas.

Apenas 24% dos passageiros da terceira classe sobreviveram, enquanto 47% se salvaram na segunda classe e 63% da primeira.

A idade média dos sobreviventes era de 29 anos.

Apenas 38% dos passageiros sobreviveram ao acideante, sendo a maior parte deles mulheres (68%).

Pessoas acompanhadas de algum familiar também foram previlegiados para embarcarem nos botes.

# Limitações

Numa primeira visão encontrou-se como fator limitante para a análise que algumas propriedades não possuiam valores para alguns dos passageiros. Estas características são: Age, Cabin, Embarked.

Medida tomada:
- Age: Os valores faltantes desta coluna foram preenchidos com base na idade média dos passageiros.
- Cabin: Como esta caracteristica não foi considerada na fase de analise optou-se por excluí-la.
- Embarked: O local de embarque foi preenchido com base na probabilidade de selecionar um dos locais de embarque da mesma classe do passageiro em questão.
